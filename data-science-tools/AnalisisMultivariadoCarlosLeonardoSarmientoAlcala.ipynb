```python
#Se inserta dataset cuyo objetivo es realizar una automatización para la aprobación o desaprobación
#de una solicitud de crédito mediante un dataset de entrada
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv("loan_data.csv", sep=';' ,encoding='latin-1')
#Se imprime un head del dataframe
print(data.head(5))
```

```python
import seaborn as sns
#Se obtienen columnas de los datos presentes
print(data.columns)
```

```python
#Visualización de estado de préstamo contra cantidad aprobada para préstamo
plt.figure(figsize = (20,5))
sns.boxplot(x='loan_status', y = 'funded_amnt',data = data, palette = ['#008B8B','#00FFFF'])
plt.show()
```

```python
#Se calculan las correlaciones entre columnas de interés
import seaborn as sns
corrmat = data[['delinq_2yrs','inq_last_6mths','mths_since_last_delinq','open_acc','pub_rec','revol_util','total_acc','loan_amnt']].corr()
plt.figure(figsize=(10,10))
sns.heatmap(corrmat)
#print(data[['loan_amnt','open_acc']])
```

```python
#Se muestran las distribuciones de la cantidad de prestamos contra el tipo de propiedad de los solicitantes
ax = sns.violinplot(x='home_ownership',y='loan_amnt',data=data)
```
